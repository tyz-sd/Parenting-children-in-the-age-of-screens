```python
def process_income_field(row):
    '''
    Processes the income string which contains range of income along with a text into a numeric value which is the average of the range
    Args:
        row (str) : Each row value of a dataframe which is a combine dincomes with a text
    Returns:
        int : Average Income of the range of values present in the row
        
    >>>>>> '$100,000 to less than $150,000'
                -> 125000
    >>>>>> 'less than $10,000'
                -> 10000
    '''
    assert isinstance(row,str),'Something is different in dataframe'
    row=row.lower().strip()
    
    if(row=='refused'):
        return 0
    else:
        income=list(map(int,re.findall(r'\d+',row)))
        total=0
        for x in income:
            total+=x
        return int(total*1000)

def pre_process_dataframe(actual,new_cols,remove_cols):
    '''
    Returns a processed dataframe which may include removing un-necessary columns, converting the income field to numeric and renaming columns
    Args:
        actaul (pd.Frame.DataFrame) : The actual dataframe which needs to be preprocessed
        new_cols (list) : List of new columns names in the order that appeared in the dataframe
    Returns:
        (pd.Frame.DataFrame) : Pre Processed dataframe
    '''
    actual.drop(remove_cols,axis=1,inplace=True)
    actual['F_INCOME']=actual['F_INCOME'].apply(lambda x: process_income_field(x))
    actual.columns=new_cols
    return actual
```
